6. Develop the code in the previous block by copying it into the next code block, and here:
   * Develop code to find out about the top 15 most ambiguous words in your adjective or verb token list, and
   * Write your output to a neatly formatted text file for reading. The text output file should provide a clear heading to provide a title for each corpus you're comparing. *(5 pts)*

```python
from nltk.corpus import wordnet as wn
import pandas as pd

def ambiguity_score(word):
    """Return the number of WordNet synsets for a word."""
    return len(wn.synsets(word))

def top_ambiguous_words(tokens, top_n=15):
    """Return a DataFrame of the top N most ambiguous words."""
    df = pd.DataFrame({"word": tokens})
    df = df.groupby("word").size().reset_index(name="freq")
    df["ambiguity"] = df["word"].apply(ambiguity_score)
    df = df[df["ambiguity"] > 0]  # keep only words WordNet recognizes
    return df.sort_values("ambiguity", ascending=False).head(top_n)

ambig_A = top_ambiguous_words(pos_A)
ambig_B = top_ambiguous_words(pos_B)

print("Top ambiguous words in Mulan corpus:")
print(ambig_A)

print("\nTop ambiguous words in transcripts corpus:")
print(ambig_B)

with open("ambiguity_results.txt", "w", encoding="utf-8") as f:
    f.write("=== Top 15 Most Ambiguous Words — Mulan Corpus ===\n")
    f.write(ambig_A.to_string(index=False))
    f.write("\n\n=== Top 15 Most Ambiguous Words — Transcripts Corpus ===\n")
    f.write(ambig_B.to_string(index=False))

print("\nAmbiguity results written to 'ambiguity_results.txt'")
```
