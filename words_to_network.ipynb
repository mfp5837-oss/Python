```python
import os
import pandas as pd

# This MUST be a folder, not a file
collPath = "."   # <-- change to your actual folder name

allDataFrames = []

# Loop through every .txt file in the folder
for file in os.listdir(collPath):
    if file.endswith(".txt"):
        filepath = f"{collPath}/{file}"
        
        with open(filepath, "r", encoding="utf8") as f:
            text = f.read()

        df = pd.DataFrame({
            "filename": [file],
            "text": [text]
        })

        allDataFrames.append(df)

# Combine all dataframes into one
finalDF = pd.concat(allDataFrames, ignore_index=True)

finalDF.head()
```

```python
import os
import pandas as pd
import spacy

nlp = spacy.load("en_core_web_lg")

# Your .txt files are in the same folder as your notebook
collPath = "."

# ---------------------------------------------------------
# WORKING wordCollector() FUNCTION
# ---------------------------------------------------------
def wordCollector(doc, episode_name): words = [] pos_tags = [] lemmas = []

    for token in doc:
        if token.is_alpha:  words.append(token.text) pos_tags.append(token.pos_) lemmas.append(token.lemma_)

    df = pd.DataFrame({"episode": episode_name,"word": words,"lemma": lemmas,"pos": pos_tags})

    return df

allDataFrames = []

# Loop through every .txt file in the folder
for file in os.listdir(collPath):
    if file.endswith(".txt"):
        filepath = f"{collPath}/{file}"
        episode_name = file.replace(".txt", "")

        with open(filepath, "r", encoding="utf8") as f:
            text = f.read()

        doc = nlp(text)
        df = wordCollector(doc, episode_name)
        allDataFrames.append(df)

finalDF = pd.concat(allDataFrames, ignore_index=True)

finalDF.head()
```
