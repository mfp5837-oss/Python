```python
import re
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def generate_word_cloud(filename):
    try:
        # 1. Load local text file
        with open(filename, "r", encoding="utf-8") as f:
            text = f.read()

        # 2. Tokenize
        words = re.findall(r"[A-Za-z']+", text.lower())

        # 3. Filter stopwords
        stopwords = {
            "the","and","to","of","a","in","that","is","it","for","on","with",
            "as","at","but","be","this","you","i"
        }
        filtered = [w for w in words if w not in stopwords and len(w) > 2]

        # 4. Generate WordCloud
        wc = WordCloud(
            width=800,
            height=800,
            background_color="white",
            stopwords=stopwords,
            colormap="spring",
            min_font_size=10
        ).generate(" ".join(filtered))

        # 5. Display the WordCloud
        plt.figure(figsize=(8, 8))
        plt.imshow(wc)
        plt.axis("off")
        plt.tight_layout(pad=0)
        plt.show()

    except FileNotFoundError:
        print("The file was not found. Please check the file path.")

# Run it
generate_word_cloud("1x2.txt")
```
