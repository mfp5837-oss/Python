```python
import requests
import matplotlib.pyplot as plt
from collections import Counter
import re

# 1. Download text
url = "https://newtfire.org/courses/tutorials/mulan1998.txt"
text = requests.get(url).text

# 2. Clean and tokenize
words = re.findall(r"[A-Za-z']+", text.lower())

# Remove common stopwords
stopwords = {"the","and","to","of","a","in","that","is","it","for","on","with","as","at","but","be","this","you","i"}
filtered = [w for w in words if w not in stopwords and len(w) > 2]

# 3. Count top words
counts = Counter(filtered).most_common(15)

labels = [w for w, c in counts]
values = [c for w, c in counts]

# 4. Plot
plt.figure(figsize=(10,5))
plt.bar(labels, values, color="orchid")
plt.title("Most Frequent Words in Mulan (1998) Transcript")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
