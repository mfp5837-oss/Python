```python
# ---------------------------------------------------------
# Imports
# ---------------------------------------------------------
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

# ---------------------------------------------------------
# One-time NLTK downloads (safe to run every time)
# ---------------------------------------------------------
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')   # NEW
nltk.download('averaged_perceptron_tagger')       # fallback

# ---------------------------------------------------------
# Load your Mulan text file
# --------------------------------------------------------
filepath = 'mulan1998_scraped.txt'

with open(filepath, 'r', encoding='utf8') as f:
    text = f.read()

tokenList = word_tokenize(text)
print("First 50 tokens:", tokenList[:50])


# Tag all tokens
tagged_all = pos_tag(tokenList)
print("\nFirst 50 POS tags (all tokens):", tagged_all[:50])

# Tag unique tokens
uniqueTokens = list(set(tokenList))
tagged_unique = pos_tag(uniqueTokens)
print("\nFirst 50 POS tags (unique tokens):", tagged_unique[:50])
```
